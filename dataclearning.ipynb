```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the uploaded file
file_path = 'training.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
data.head()
data.shape
```

```python
# Dropping rows where 'OilPeakRate' is NaN
cleaned_data = data.dropna(subset=['OilPeakRate'])

# Saving the cleaned data to a new file
cleaned_file_path = 'cleaned_training_data.csv'
cleaned_data.to_csv(cleaned_file_path, index=False)

cleaned_data.shape
```

```python
# Calculate the threshold for dropping rows (50% of the total number of columns)
threshold = len(cleaned_data.columns) * 0.5

# Dropping rows with more than 50% missing values
cleaned_data_with_threshold = cleaned_data.dropna(thresh=threshold)

# Saving the further cleaned data to a new file
further_cleaned_file_path = 'further_cleaned_training_data.csv'
cleaned_data_with_threshold.to_csv(further_cleaned_file_path, index=False)

cleaned_data_with_threshold.shape
```

```python
# Calculate the percentage of missing data in each column
missing_percentage = cleaned_data_with_threshold.isnull().mean() * 100

# Identify columns where the percentage of missing data is greater than 80%
columns_to_drop = missing_percentage[missing_percentage > 80].index

# Drop these columns from the DataFrame
df_droped_column = cleaned_data_with_threshold.drop(columns=columns_to_drop)

# Save the modified DataFrame back to a CSV file
cleaned_file_path = 'cleaned_further_cleaned_training_data.csv'
df_droped_column.to_csv(cleaned_file_path, index=False)
df_droped_column = df_droped_column.drop(columns=['frac_type'])

# Return the path of the cleaned file and the dropped columns for reference
cleaned_file_path, columns_to_drop.tolist()
df_droped_column.shape
```

```python


# Function to plot distribution of each column
def plot_column_distributions(df):
    for column in df.columns:
        plt.figure(figsize=(10, 4))

        # Check if the column is numeric or categorical
        if pd.api.types.is_numeric_dtype(df[column]):
            # Plot histogram for numeric columns
            sns.histplot(df[column], kde=True, bins=30)
            plt.title(f'Histogram of {column}')
        else:
            # Plot bar plot for categorical columns
            value_counts = df[column].value_counts()
            sns.barplot(x=value_counts.index, y=value_counts)
            plt.title(f'Bar Plot of {column}')
            plt.xticks(rotation=45)

        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.tight_layout()
        plt.show()

# Plotting distributions for each column
plot_column_distributions(data)
```

```python
# Calculating skewness for each numeric column in the dataset
skewness_values = data.select_dtypes(include=['float64', 'int64']).skew()

# Displaying the skewness values
skewness_values.sort_values(inplace=True)
skewness_values
```

```python
# Threshold for considering a column as significantly skewed
skewness_threshold = 1.0

# Impute missing values: Median for significantly skewed columns, mean for others
for column in df_droped_column.columns:
    if pd.api.types.is_numeric_dtype(df_droped_column[column]):
        if abs(df_droped_column[column].skew()) > skewness_threshold:
            # Impute with median for significantly skewed columns
            df_droped_column[column].fillna(df_droped_column[column].median(), inplace=True)
        else:
            # Impute with mean for other columns
            df_droped_column[column].fillna(df_droped_column[column].mean(), inplace=True)

# Check if there are any missing values left
missing_values_after_imputation = df_droped_column.isnull().sum()
missing_values_after_imputation[missing_values_after_imputation > 0]
```

```python


# Imputing missing values in 'relative_well_position', 'batch_frac_classification', and 'well_family_relationship' with their most frequent types (mode)
for column in ['ffs_frac_type', 'relative_well_position', 'batch_frac_classification', 'well_family_relationship']:
    mode_value = df_droped_column[column].mode()[0]
    df_droped_column[column].fillna(mode_value, inplace=True)

# Check again for any remaining missing values in these columns
missing_values_final_check_specific_columns = df_droped_column[['relative_well_position', 'batch_frac_classification', 'well_family_relationship']].isnull().sum()
missing_values_final_check_specific_columns
# # Define the path for the exported CSV file
exported_file_path = 'cleaned_and_imputed_data.csv'

# Exporting the data to a CSV file
df_droped_column.to_csv(exported_file_path, index=False)

```
