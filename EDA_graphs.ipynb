```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the uploaded file
file_path = 'cleaned_and_imputed_data 2.csv'
data = pd.read_csv(file_path)
```

```python
# Set the aesthetic style of the plots
sns.set_style('whitegrid')
sns.set_context('talk')  # This makes elements like labels larger

# Set up the figure
plt.figure(figsize=(8, 6))

# Create the histogram with a kernel density estimate
sns.histplot(data['OilPeakRate'], kde=True, color='skyblue', edgecolor='black')

# Set the titles and labels
plt.title('Distribution of Oil Peak Production Rate', fontsize=20)
plt.xlabel('Oil Peak Rate', fontsize=16)
plt.ylabel('Frequency', fontsize=16)

# Remove the top and right spines
sns.despine()

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
data_num = data.select_dtypes(exclude=['object'])
del data_num['Unnamed: 0']
plt.figure(figsize=(10, 10))
sns.heatmap(data_num.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(15, 12))

# Create a histogram for each numerical column
for i, column in enumerate(data_num.columns):
    plt.subplot((len(data_num.columns) + 2) // 3, 3, i + 1)
    sns.histplot(data_num[column], kde=False, color="skyblue")
    plt.title(column)
    plt.tight_layout()

plt.show()
```
